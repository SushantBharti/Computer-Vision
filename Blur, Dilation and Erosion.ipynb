````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Bluring images

````python
path=r'..\Attach'

image=cv2.imread(r"butterfly.jpg")

blur=cv2.GaussianBlur(image,(5,55),0)

cv2.imwrite(path+r"\blur.png",blur)
plt.figure(figsize=(13,15))
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(blur)
````

````python
cv2.imshow('Blur',np.concatenate([image, blur],axis=1))
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
image=cv2.imread(path+r'\wellner.png')
````

````python
kernel=np.ones((4,4),'uint8')
dilate=cv2.dilate(image,kernel,iterations=1)
erode=cv2.erode(image,kernel,iterations=1)
````

````python
plt.figure(figsize=(13,15))

plt.subplot(1,3,1)
plt.imshow(image)

plt.subplot(1,3,2)
plt.imshow(dilate)
plt.subplot(1,3,3)
plt.imshow(erode)
plt.savefig(path+r'\BlurDilateErosion.png')
plt.show()
````
