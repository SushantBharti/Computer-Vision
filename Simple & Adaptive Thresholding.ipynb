```python
binary=np.zeros([ht,wt,3],'uint8')
```

```python
thresh=30
for row in range(0,ht):
    for col in range(0,wt):
        if bw[row][col].min()>thresh:
            binary[row][col]=[255,0,0]
        else:
            if bw[row][col].max()>thresh:
                binary[row][col]=[0,255,0]
```

## Adaptive Thresholding

```python
img=cv2.imread(path+r"\sudoku.jpg",0)
#src=cv2.imread(path+r"\sudoku.jpg")
```

```python
ret, thresh_basic= cv2.threshold(img,70,255,cv2.THRESH_BINARY)
thres_adapt=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,115,1)
```

```python

plt.figure(figsize=(14,7))
plt.subplot(131)
plt.imshow(img)
plt.subplot(132)
plt.imshow(thresh_basic)
plt.subplot(133)
plt.imshow(thres_adapt)
#plt.savefig()
```

```python
thres_adapt.shape,img.shape,thresh_basic.shape
```

```python
cv2.imwrite(path+r"\sudokuThresh.png",np.concatenate((img,thresh_basic,thres_adapt),axis=1))
```
