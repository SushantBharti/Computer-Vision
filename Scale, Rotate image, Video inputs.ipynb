```python
image=cv2.imread(r"butterfly.jpg")
```

```python
img_half=cv2.resize(image,(0,0), fx=0.5,fy=0.5)
#plt.figure(figsize=(13,15))
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(img_half)
```

```python
img_stretch=cv2.resize(image,(200,200))
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(img_stretch)
```

```python
img_stretch_near=cv2.resize(image,(200,200),interpolation=cv2.INTER_NEAREST)
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(img_stretch_near)
```

```python
ht,wt,ch=image.shape
```

```python
M=cv2.getRotationMatrix2D((ht/2,wt/2),-30,1)
```

```python
rotated=cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(rotated)
```

```python
cap=cv2.VideoCapture(path+r'\bunny.mp4')
cap
```

```python
from matplotlib import animation, rc
```

```python
count=0
while(count<10):
    ret,frame=cap.read()
    #print(type(ret),ret)
    frame= cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    #plt.show()
    count=count+1
    cv2.imshow("Frame",frame)
cap.release()
```

```python
from IPython.display import HTML
def animate():
    ret,frame=cap.read()
    frame= cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    cap.release()
    return frame
```

```python
fig, ax = plt.subplots()
anim=animation.FuncAnimation(fig,animate , 
                               frames=100, interval=20, blit=True)
HTML(anim.to_html5_video())
```
