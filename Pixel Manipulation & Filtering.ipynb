```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path=r'E:\Forecasting Demand\Web Service\Attach'
```

```python
color=cv2.imread(r"butterfly.jpg")
```

```python
color.shape
```

```python
ht,wt,ch=color.shape
```

```python
b,g,r=cv2.split(color)
```

```python
rgb_split=np.empty([ht,wt*3,3],'uint8')
rgb_split[:,0:wt]=cv2.merge([b,b,b])
rgb_split[:,wt:wt*2]=cv2.merge([g,g,g])
rgb_split[:,wt*2:]=cv2.merge([r,r,r])
cv2.imwrite(path+r"\rgb_split.png",rgb_split)
```

```python
hsv=cv2.cvtColor(color,cv2.COLOR_BGR2HSV)

h,s,v=cv2.split(hsv)
hsv_split=np.concatenate((h,s,v),axis=1)
cv2.imwrite(path+r"\hsv_split.png",hsv_split)
```
